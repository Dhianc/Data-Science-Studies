## Seleções

```python
# Imoveis classificados com tipo 'Apartamento'

selecao_apartamento = dados_residencial['Tipo'] == 'Apartamento'
s1 = dados_residencial[selecao_apartamento].shape[0]

print('Nº de imóveis classificados com tipo "Apartamento": ', s1)
```

```python
# Imóveis classificados com tipos "Casa", 'Casa de Condomínio' e 'Casa de vila' 

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila' ]
selecao_casa = dados_residencial['Tipo'].isin(casa)
s2 = dados_residencial[selecao_casa].shape[0]
print("Nº de imóveis classificados com tipos 'Casa', 'Casa de Condominio' e 'Casa de Vila': ", s2)
```

```python
# Imóveis com área entre 60 e 100 metros quadrados, incluindo limites

selecao_area = (dados_residencial['Area'] >= 60) & (dados_residencial['Area'] <= 100)
s3 = dados_residencial[selecao_area].shape[0]
print("Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites: ", s3)
```

```python
# Imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00

selecao4 = (dados_residencial['Quartos'] >= 4) & (dados_residencial['Valor'] < 2000)
s4 = dados_residencial[selecao4].shape[0]
print("Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que 2 mil: ", s4)
```

## Tratamento de dados faltantes

```python
dados_residencial.dropna(subset = 'Valor', inplace = True)
dados_residencial.info()
```

```python
selecao_apartamento_condominio_nulo = (dados_residencial['Tipo'] == 'Apartamento') & (dados_residencial['Condominio'].isnull())
dados_residencial = dados_residencial[~selecao_apartamento_condominio_nulo]
dados_residencial.fillna(0, inplace=True)
dados_residencial.info()

# Salvar novos dados residenciais sem dados NaN
```

## Novas variáreis

```python
dados_residencial['Valor Bruto'] = dados_residencial['Valor'] + dados_residencial['Condominio'] + dados_residencial['IPTU']
dados_residencial['Valor m2'] = (dados_residencial['Valor'] / dados_residencial['Area']).round(2)
dados_residencial['Valor Bruto m2'] = (dados_residencial['Valor Bruto'] / dados_residencial['Area']).round(2)
dados_residencial['Tipo Agregado'] = dados_residencial['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')

dados_residencial
```

## Excluindo variáreis

```python
dados_residencial.drop(['Valor Bruto', 'Valor Bruto m2'], axis=1, inplace=True)
dados_residencial
```

## Criando agrupamentos

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao_bairros = dados_residencial['Bairro'].isin(bairros)
dados_residencial_group = dados_residencial[selecao_bairros]
dados_residencial_group['Bairro'].drop_duplicates(inplace=True)
grupo_bairro = dados_residencial_group.groupby('Bairro')
grupo_bairro['Valor', 'Condominio'].mean().round(2)
```

## Identificando e removendo outliers

```python
plt.rc('figure', figsize = (14, 6))
```

```python
dados_residencial.boxplot(['Valor'])
```

```python
dados_residencial[dados_residencial['Valor'] >= 500000]
```

```python
valor = dados_residencial['Valor']
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >= limite_inferior) & (valor <= limite_superior)
dados_new = dados_residencial[selecao]
dados_new.boxplot(['Valor'])
```

```python
dados_residencial.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = dados.groupby('Tipo')['Valor']
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados_residencial['Tipo'] == tipo
    eh_dentro_limite = (dados_residencial['Valor'] >= limite_inferior[tipo]) & (dados_residencial['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados_residencial[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```

```python
dados_new.boxplot(['Valor'], by = ['Tipo'])
```
